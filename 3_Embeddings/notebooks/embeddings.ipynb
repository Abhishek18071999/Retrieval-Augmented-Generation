```python
loader = PyPDFLoader("../knowledge_base/Openclaw_Research_Report.pdf")
docs = loader.load()
len(docs)
```

```python
chunker = RecursiveCharacterTextSplitter(chunk_size=300, chunk_overlap=50)
chunks = chunker.split_documents(docs)
len(chunks)
```

```python
# embed documents
text_documents = [chunk.page_content for chunk in chunks]
print(len(text_documents))
```

```python
# embed documents

document_embeddings = ollama.embed(
    model="embeddinggemma",
    input=text_documents
)
```
