## json -- loader

```python
from langchain_community.document_loaders.json_loader import JSONLoader
from pprint import pp
from pathlib import Path
```

```python
## Creating the path of the file
file_path = Path("../Knowledge_base/apparels.json");
file_path.exists()
```

```python
def metadata_func(record:dict,metadata:dict) -> dict:
    metadata["product_name"] = record["productName"]
    metadata["category"] = record["category"]
    metadata["manufacturer"] = record["manufacturer"]

    ## delete a key -value pair from json structure
    del metadata["seq_num"]
    return metadata
```

```python
## loading the json file
loader = JSONLoader(file_path=file_path.as_posix(), jq_schema=".products[]",
                    content_key="Description",metadata_func=metadata_func,)
```

```python
docs = loader.load()
```
