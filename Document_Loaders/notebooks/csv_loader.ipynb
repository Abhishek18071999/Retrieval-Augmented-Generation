## Loader -> csv Loader

```python
from langchain_community.document_loaders.csv_loader import CSVLoader
from pprint import pp
from pathlib import Path
```

```python
file_path = Path("../knowledge_base/organizations.csv")
```

```python
loader = CSVLoader(file_path=file_path.as_posix(),metadata_columns=['Description'],source_column='Industry',
                   content_columns=['Description']) 
```

```python
docs = loader.load()
```
