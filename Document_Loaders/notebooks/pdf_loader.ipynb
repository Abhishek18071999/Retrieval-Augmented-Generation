## Loader -> PyPDF


```python
from langchain_community.document_loaders import PyPDFLoader
from pprint import pp
from pathlib import Path
```

```python
## creating the path for the file
file_path = Path("../knowledge_base/NIPS-2017-attention-is-all-you-need-Paper.pdf")
file_path.as_posix()
```

```python
## create the loader
pdfLoader = PyPDFLoader(file_path=file_path.as_posix(), mode = 'page')
```

```python
## load the pdf
documents = pdfLoader.load()
```

```python
len(documents)
```

```python
print(documents[0].page_content)
```

```python
from langchain_community.document_loaders.parsers  import TesseractBlobParser, RapidOCRBlobParser
```

```python
## Extracting images from pdf using parsers
pypdf_image_loader = PyPDFLoader(file_path=file_path.as_posix(),mode='page',extract_images=True, 
                                images_parser=RapidOCRBlobParser(),
                                images_inner_format='html-img' 
                                 )
```

```python
documents_with_images = pypdf_image_loader.load()
```

```python
len(documents_with_images)
```

```python
page_with_image = documents_with_images[2]
```

```python
print(page_with_image)
```
