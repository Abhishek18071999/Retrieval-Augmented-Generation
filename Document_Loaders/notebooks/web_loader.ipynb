## loader --> WEB_LOADER

```python
from langchain_community.document_loaders import WebBaseLoader
from pprint import pp
```

```python
url = "https://reference.langchain.com/python/langchain/overview";
loader = WebBaseLoader(web_path=url)
```

```python
docs = loader.load()
```

```python
len(docs)
```

## Recursive Loader

```python
from langchain_community.document_loaders import RecursiveUrlLoader
```

```python
base_url="https://reference.langchain.com/python/langchain"
loader = RecursiveUrlLoader(url=base_url, max_depth=2)
```

```python
docs = loader.load()
```

```python
len(docs)
```

```python
for i in range(10):
    doc = docs[i];
    print(doc.metadata, end = "\n\n")
```

## LAZY LOADING

```python
lazy_loader = loader.lazy_load()
```

```python
print(lazy_loader)
```

```python
counter= 0
for i in lazy_loader:
    if counter == 10:
        break
    counter+=1
    pp(i.page_content[0:300] +"\n\n")
```
